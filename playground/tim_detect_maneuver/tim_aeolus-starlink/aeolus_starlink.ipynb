```python
import pandas as pd
```

```python
df = pd.read_csv("aeolus_starlink.csv")
len(df)
```

Some entries are duplicated

```python
df.drop_duplicates(subset=['EPOCH','NORAD_CAT_ID'], inplace=True)
len(df)
```

```python
df['EPOCH'] = df['EPOCH'].astype('datetime64[ns]')
df = df.set_index("EPOCH")
```

* Only look at some columns.
* Convert EPOCH to datetime
* Set index to EPOCH

```python
df = df[["NORAD_CAT_ID","MEAN_MOTION","ECCENTRICITY","INCLINATION","ARG_OF_PERICENTER","MEAN_ANOMALY",
    "BSTAR","MEAN_MOTION_DOT","SEMIMAJOR_AXIS","PERIOD","APOAPSIS","PERIAPSIS"]]
df
```

Can you tell when AEOLUS maneuvered? It was around Sept 2nd.  Hard to tell since AEOLUS seem to move around a lot.
