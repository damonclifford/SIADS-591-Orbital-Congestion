```python
test = df[30000:32500].copy()
test = test[['SEMIMAJOR_AXIS','INCLINATION']]
test['SEMIMAJOR_AXIS_DIFF'] = test.SEMIMAJOR_AXIS.diff()
test['ROLLING'] = test.SEMIMAJOR_AXIS.rolling(5).median() # remove outliers
test['ROLLING_DIFF'] = test.ROLLING.diff()
std3 = test['ROLLING_DIFF'].std()*2
events = test[~test.ROLLING_DIFF.between(-std3,std3)].dropna()



fig, axs = plt.subplots(2, 1)
test.SEMIMAJOR_AXIS.plot(ax=axs[0],figsize=(10,8))
test['ROLLING'].plot(ax=axs[1])

events
```
